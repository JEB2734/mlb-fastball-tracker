```python
from pybaseball import statcast
import pandas as pd

# Pull one month (July 2024) as a test
df = statcast(start_dt="2024-07-01", end_dt="2024-07-31")

# Keep only four-seam fastballs and the columns we'll actually use downstream
cols = ["game_date", "player_name", "pitcher", "pitch_type",
        "release_speed", "home_team", "away_team", "inning_topbot"]
ff = df[df["pitch_type"] == "FF"][cols].dropna(subset=["release_speed"])

# Save it to a CSV
ff.to_csv("fastballs_2024.csv", index=False)

print(f"Saved {len(ff):,} four-seam pitches to fastballs_2024.csv")
ff.head()
```

```python
from pybaseball import statcast, cache
import pandas as pd

cache.enable()  # makes re-runs fast and protects finished seasons if it hiccups

cols = ["game_date", "player_name", "pitcher", "pitch_type",
        "release_speed", "home_team", "away_team", "inning_topbot"]

all_seasons = []
for year in range(2015, 2026):          # 2015 through 2025
    print(f"Pulling {year}...", end=" ", flush=True)
    df = statcast(start_dt=f"{year}-07-01", end_dt=f"{year}-07-31")
    ff = df[df["pitch_type"] == "FF"][cols].dropna(subset=["release_speed"]).copy()
    ff["season"] = year
    all_seasons.append(ff)
    print(f"{len(ff):,} four-seam pitches")

fastballs = pd.concat(all_seasons, ignore_index=True)
fastballs.to_csv("fastballs_all.csv", index=False)

print(f"\nDone. Saved {len(fastballs):,} pitches across "
      f"{fastballs['season'].nunique()} seasons to fastballs_all.csv")
```
